```python
#https://neptune.ai/blog/arima-sarima-real-world-time-series-forecasting-guide
def check_stationarity(ts):
    dftest = adfuller(ts)
    adf = dftest[0]
    pvalue = dftest[1]
    critical_value = dftest[4]['5%']
    if (pvalue < 0.05) and (adf < critical_value):
        print('The series is stationary')
    else:
        print('The series is NOT stationary')
```

#ARIMA BY INSTITUTION

```python
def find_institution_id(institution_name, institution_df):
    """
    Find the institution ID based on the institution name.

    Parameters:
        institution_name (str): Name of the institution to search for.
        institution_df (DataFrame): DataFrame containing institution IDs and names.

    Returns:
        institution_id (str): ID of the institution found (or None if not found).
    """
    institution_df = institution_df.copy()
    institution_id = institution_df[institution_df['INSTNM'] == institution_name]['UNITID'].values
    if len(institution_id) > 0:
        return institution_id[0]
    else:
        print(f"Institution '{institution_name}' not found.")
        return None
```

```python
def predict_enrollment_2022(institution_name, enrollment_data, institution_df):
    """
    Predict enrollment for the year 2022 for a specific institution.

    Parameters:
        institution_name (str): Name of the institution to predict enrollment for.
        enrollment_data (DataFrame): DataFrame containing enrollment data.
        institution_df (DataFrame): DataFrame containing institution IDs and names.

    Returns:
        metrics (dict): Dictionary containing evaluation metrics for the prediction.
    """
    # Find the institution ID based on the institution name
    institution_id = find_institution_id(institution_name, institution_df)

    if institution_id is not None:
        # Get data for the specified institution
        institution_data = enrollment_data[enrollment_data['UNITID'] == institution_id].copy()

        # Convert 'year' column to datetime format using .loc
        institution_data.loc[:, 'YEAR'] = pd.to_datetime(institution_data['YEAR'].astype(str))


        # Split data into train and test sets (using data up to 2022 for training)
        train_years = institution_data['YEAR'].dt.year < 2022
        test_years = institution_data['YEAR'].dt.year == 2022
        train_data = institution_data[train_years]['EFDEEXC']
        test_data = institution_data[test_years]['EFDEEXC']

        # Use auto_arima to auto select the optimal ARIMA model parameters
        model = auto_arima(train_data, seasonal=False, trace=True)

        # Fit ARIMA model with the selected parameters on the training data
        model_fit = model.fit(train_data)

        # Forecast enrollment for 2022
        forecast_2022 = model_fit.predict(n_periods=len(test_data))

        plt.bar(['Actual', 'Predicted'], [test_data.values[0], forecast_2022.values[0]])
        plt.xlabel('Data Type')
        plt.ylabel('Enrollment')
        plt.title(f'Actual vs Predicted Enrollment for {institution_name} in 2022')
        plt.grid(False)
        plt.show()

        print (test_data.values)
        print (forecast_2022.values)



        # Calculate evaluation metrics
        mae = np.mean(np.abs(test_data.values - forecast_2022))
        mse = np.mean((test_data.values - forecast_2022) ** 2)
        rmse = np.sqrt(mse)

        # Create a dictionary to store evaluation metrics
        metrics = {
            'MAE': mae,
            'MSE': mse,
            'RMSE': rmse
        }

        return metrics

    else:
        print(f"No data found for institution '{institution_name}'.")
        return None
```

```python

from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense

# Select data from 2018 to 2021 for training
train_data = aggregate_data2[aggregate_data2.index.year < 2022]
test_data = aggregate_data2[aggregate_data2.index.year == 2022]

# Scale the data
scaler = MinMaxScaler()
scaled_train_data = scaler.fit_transform(train_data.iloc[:, 1:])  # Exclude YEAR col

# Define the number of time steps (use past 3 years' data to predict 2022)
n_steps = 3

# Prepare the training data
X_train, y_train = [], []
for i in range(n_steps, len(train_data)):
    X_train.append(scaled_train_data[i - n_steps:i, :])
    y_train.append(scaled_train_data[i, 0])

X_train, y_train = np.array(X_train), np.array(y_train)

if len(X_train) == 0:
    print("Error: No data found for training. Check train_data or n_steps value.")
else:
    print("Successfully created X_train and y_train.")

# Build the LSTM model
model = Sequential([
    LSTM(units=50, activation='relu', input_shape=(X_train.shape[1], X_train.shape[2])),
    Dense(units=1)
])

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, epochs=100, batch_size=32)

# Predict EFDEEXC for the year 2022
# Use past 3 years' data (2019, 2020, 2021) to predict 2022
past_years_data = aggregate_data2.iloc[-n_steps:, 1:].values  # Exclude the YEAR column
scaled_past_years_data = scaler.transform(past_years_data)

# Reshape the input data to match the expected shape
reshaped_past_years_data = scaled_past_years_data.reshape(1, n_steps, 2)

# Predict EFDEEXC for 2022
predicted_efdeexc_2022_scaled = model.predict(reshaped_past_years_data)

# # Perform inverse transformation
predicted_efdeexc_2022 = scaler.inverse_transform(predicted_efdeexc_2022_scaled)

print("Predicted EFDEEXC for 2022:", predicted_efdeexc_2022)
```

```python
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense

# Select relevant features and convert to numpy array
data = aggregate_data2[['EFDEEXC', 'EFDESOM', 'EFDENON']].values

# Normalize the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

# Define the number of time steps for the LSTM model
n_steps = 4

# Split data into input and output
X, y = [], []
for i in range(len(scaled_data) - n_steps):
    X.append(scaled_data[i:i + n_steps])
    y.append(scaled_data[i + n_steps, 0])  # We are predicting EFDEEXC, so take only the first column

X, y = np.array(X), np.array(y)

# Reshape input data for LSTM (samples, time steps, features)
n_features = X.shape[2]
X = X.reshape((X.shape[0], X.shape[1], n_features))

# Define the LSTM model
model = Sequential()
model.add(LSTM(units=50, activation='relu', input_shape=(n_steps, n_features)))
model.add(Dense(units=1))  # Output layer, one neuron for prediction
model.compile(optimizer='adam', loss='mse')

# Fit the model
model.fit(X, y, epochs=200, verbose=1)

# Forecast
forecast_input = scaled_data[-n_steps:].reshape((1, n_steps, n_features))
forecast = model.predict(forecast_input)

# Inverse transform the forecasted value to get the actual scale
forecast = scaler.inverse_transform(np.hstack([forecast, np.zeros((forecast.shape[0], 2))]))[0, 0]
rounded_forecast = round(forecast, 2)
print("Forecasted EFDEEXC for year 2023:", rounded_forecast)
```
